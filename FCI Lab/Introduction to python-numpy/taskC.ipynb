Data Generation

```python
from numpy.random import rand, randn
```

```python
n, d, k = 100, 2, 2
```

```python
np.random.seed(20)
X = rand(n, d)

# means = [rand(d)  for _ in range(k)]  # works for any k
means = [rand(d) * 0.5 + 0.5 , - rand(d)  * 0.5 + 0.5]  # for better plotting when k = 2

S = np.diag(rand(d))

sigmas = [S]*k # we'll use the same Sigma for all clusters for better visual results

print(means)
print(sigmas)
```

Solution
===

```python
def compute_log_p(X, mean, sigma):
    ''' fill your code in here...

    '''
    d = len(mean)
    # Calculate the determinant and inverse of Sigma
    Sigma_inv = np.linalg.inv(sigma)
    log_det_Sigma = np.log(np.linalg.det(sigma))

    # Calculate the quadratic form
    diff = X - mean
    quad_form = diff @ Sigma_inv @ diff.T

    # Calculate the log PDF
    log_pdf = -0.5 * (d * np.log(2 * np.pi) + log_det_Sigma + quad_form)

    return log_pdf
```

```python
log_ps = [compute_log_p(X, m, s) for m, s in zip(means, sigmas)]  # exercise: try to do this without looping
```

```python
assignments = np.argmax(log_ps, axis=0)
print(assignments)
```

```python
colors = np.array(['red', 'green'])[assignments[:,0]]
plt.scatter(X[:, 0], X[:, 1], c = colors, s=100)
plt.scatter(np.array(means)[:, 0], np.array(means)[:, 1], marker='*', s=200)
plt.show()
```
